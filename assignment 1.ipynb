6.Write a function to demonstrate the behavior of mutable and immutable arguments.


```python
def demo(num, text):
    num = num + 5          # int (immutable)
    text = text + " World" # string (immutable)
    print("Inside:", num, text)

x = 10
s = "Hello"

demo(x, s)

print("Outside:", x, s)
```
